```python
# Data Loading 
import matplotlib.pyplot as plt
import json
import pandas as pd
%matplotlib inline

with open('data_2010-2019.json', 'r') as f:
    data = json.load(f)
df = pd.DataFrame(data)

# Useless columns removing
df = df.drop(columns=['link','logo_home','logo_away'])

# Columns Sorting
df = df[['saison_annee','journee ','date','time','home','home_score','away_score','away','score']]
df.head()
```

```python
# Match suivant une saison 
def getTeamSeason( equipe,  saison,terrain="all"):
    
    if terrain =="home":
        terrain_condition = (df.home==equipe)  
    elif terrain=="away":
        terrain_condition =  (df.away==equipe) 
    else:
        terrain_condition =  (df.home==equipe) | (df.away==equipe )

    return df[( terrain_condition) & (df.saison_annee==saison)]
```

```python
# Confrontation entre deux équipes
def getOppositions(team1, team2, terrain="all"):
    """
    team1, team2 : les équipes 
    terrain : "team1" for team1 games at home, "team2" for team2 games at home, "all"  for indifference
    """
    
    if terrain=="team1":
        terrainCondition= ((df.home==team1) & (df.away==team2))
    elif terrain=="team2":
        terrainCondition= ((df.home==team2) & (df.away == team1))
    else :
        terrainCondition = ((( df.home == team1)  & ( df.away == team2) ) | ( (df.home == team2) & (df.away == team1) ))
        
    return df[terrainCondition]    
```

## Score difference

```python
def getScoreDifference(ecart, team1="all", team2="all"):
    """
     ecart: difference between teams
     team1: team 1 involved
     team2 : team 2 involved
    """
    team1Condition = True
    team2Condition = True
    if team1!= "all":
        team1Condition = ( (df.home == team1) | (df.away == team1))
    if team2!= "all":    
        team2Condition = ((df.home == team2) | (df.away == team2))
        
    return df[team1Condition & team2Condition & (abs(df.home_score.astype(int) - df.away_score.astype(int)) >= ecart)]
```

## Visualising season Running

```python
def season_running(season, team1, team2):
    df_season=df[df.saison_annee==season]
    Teams= df_season.home.unique()

    TeamLists= {Team : [0] for Team in Teams}
    #For each row in our dataframe, I want to do the following:
    for row in df_season.itertuples():
        #Add the home and away team names to the correct variable
        Home = row.home
        Away = row.away

        #If the home team goals (FTHG column in the dataframe) are higher than the away team, give the correct points to each team
        if row.home_score > row.away_score:
            TeamLists[Home].append(TeamLists[Home][-1]+3)
            TeamLists[Away].append(TeamLists[Away][-1]+0)
        #If the home team goals are less than the away team, give the correct points
        elif row.home_score < row.away_score:
            TeamLists[Home].append(TeamLists[Home][-1]+0)
            TeamLists[Away].append(TeamLists[Away][-1]+3)
        #In any other case (a draw), give the correct points
        else:
            TeamLists[Home].append(TeamLists[Home][-1]+1)
            TeamLists[Away].append(TeamLists[Away][-1]+1)
            

   
    Matchday = list(range(0,39))
    #Create the bare bones of what will be our visualisation
    fig, ax = plt.subplots()

    #Add our data as before, but setting colours and widths of lines
    plt.plot(Matchday, TeamLists[team1], color = "#6CABDD", label=team1,linewidth=2)
    plt.plot(Matchday, TeamLists[team2], color = "#231F20", label=team2,linewidth=2)


    #Give the axes and plot a title each
    plt.xlabel('Gameweek')
    plt.ylabel('Points')
    plt.title(team1+'-'+team2+' Running Points')

    #Add a faint grey grid
    plt.grid()
    ax.xaxis.grid(color = "#F8F8F8")
    ax.yaxis.grid(color = "#F9F9F9")


    leg = ax.legend()
    #Remove the margins between our lines and the axes
    plt.margins(x=0,y=0)

    #Remove the spines of the chart on the top and right sides
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    
    return ax
```

```python
# Dataframe Styling
def highlight_max(data, color='yellow'):
    '''
    highlight the maximum in a Series or DataFrame
    '''
    attr = 'background-color: {}'.format(color)
    if data.ndim == 1:  # Series from .apply(axis=0) or axis=1
        is_max = data == data.max()
        return [attr if v else '' for v in is_max]
    else:  # from .apply(axis=None)
        is_max = data == data.max().max()
        return pd.DataFrame(np.where(is_max, attr, ''),
                            index=data.index, columns=data.columns)
    
def season_standings(season):
    df_season=df[df.saison_annee==season]
    Teams= df_season.home.unique()
    TeamLists= {Team : {'point':0,'ratio':0,  'J':0, 'G':0, 'N':0,  'P':0, 'p':0, 'c':0, 'dif':0} for Team in Teams}
    TeamLists['LOSC']['point']

    #For each row in our dataframe, I want to do the following:
    for row in df_season.itertuples():
        #Add the home and away team names to the correct variable
        Home = row.home
        Away = row.away

        #If the home team goals (FTHG column in the dataframe) are higher than the away team, give the correct points to each team
        if row.home_score > row.away_score:
            TeamLists[Home]['point'] += 3
            TeamLists[Home]['G']+=1 
            TeamLists[Away]['P']+=1 
        #If the home team goals are less than the away team, give the correct points
        elif row.home_score < row.away_score:

            TeamLists[Away]['point'] +=3
            TeamLists[Away]['G']+=1 
            TeamLists[Home]['P']+=1 

        #In any other case (a draw), give the correct points
        else:
            TeamLists[Home]['point'] += 1
            TeamLists[Away]['point'] += 1
            TeamLists[Home]['N']+=1         
            TeamLists[Away]['N']+=1 

        # Increment game played    
        TeamLists[Home]['J']+=1         
        TeamLists[Away]['J']+=1

        # Increment goals 
        TeamLists[Home]['p'] += int(row.home_score)
        TeamLists[Home]['c'] += int(row.away_score)
        TeamLists[Away]['p'] += int(row.away_score)
        TeamLists[Away]['c'] += int(row.home_score)

        # Increment and decrement diff
        TeamLists[Home]['dif'] = TeamLists[Home]['p'] - TeamLists[Home]['c']    
        TeamLists[Away]['dif'] = TeamLists[Away]['p'] - TeamLists[Away]['c']

        # Compute points ratio 
        TeamLists[Home]['ratio'] = float(TeamLists[Home]['point'] / TeamLists[Home]['J'])
        TeamLists[Away]['ratio'] = float(TeamLists[Away]['point'] / TeamLists[Away]['J'])
        
        classement = pd.DataFrame.from_dict(TeamLists, orient='index')
        classement = classement.sort_values(by=['point'], ascending=False)
        

    return classement
    
```
