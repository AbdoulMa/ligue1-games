```python
# Data Loading 
import json
import pandas as pd
with open('data_2010-2019.json', 'r') as f:
    data = json.load(f)
df = pd.DataFrame(data)

# Useless columns removing
df = df.drop(columns=['link','logo_home','logo_away'])

# Columns Sorting
df = df[['saison_annee','journee ','date','time','home','home_score','away_score','away','score']]
df.head()
```

## Number of years of presence in L1 between 2009 and 2019

```python
# Présence des clubs en Ligue 1
df.groupby(['home','saison_annee']).size().groupby('home').size().sort_values(ascending=False)
```

```python
# Match suivant une saison 
def getTeamSeason( equipe,  saison,terrain="all"):
    
    if terrain =="home":
        terrain_condition = (df.home==equipe)  
    elif terrain=="away":
        terrain_condition =  (df.away==equipe) 
    else:
        terrain_condition =  (df.home==equipe) | (df.away==equipe )

    return df[( terrain_condition) & (df.saison_annee==saison)]
```

##   OM  2018-2019 Season

```python
getTeamSeason(equipe="Olympique de Marseille",saison='2018/2019',terrain="all")
```

##    PSG 2015-2016 Season at Home

```python
getTeamSeason(equipe="Paris Saint-Germain",saison='2015/2016',terrain="home")
```

```python
# Confrontation entre deux équipes
def getOppositions(team1, team2, terrain="all"):
    """
    team1, team2 : les équipes 
    terrain : "team1" for team1 games at home, "team2" for team2 games at home, "all"  for indifference
    """
    
    if terrain=="team1":
        terrainCondition= ((df.home==team1) & (df.away==team2))
    elif terrain=="team2":
        terrainCondition= ((df.home==team2) & (df.away == team1))
    else :
        terrainCondition = ((( df.home == team1)  & ( df.away == team2) ) | ( (df.home == team2) & (df.away == team1) ))
        
    return df[terrainCondition]    
```

## Games  PSG vs OM between 2009 and 2019

```python
getOppositions(team1='Olympique de Marseille',team2='Paris Saint-Germain')
```

## Games Lyon vs PSG between 2009 and 2019 at Lyon

```python
getOppositions(team1='Olympique Lyonnais',team2='Paris Saint-Germain',terrain='team1')
```

## Score difference

```python
def getScoreDifference(ecart, team1="all", team2="all"):
    """
     ecart: difference between teams
     team1: team 1 involved
     team2 : team 2 involved
    """
    team1Condition = True
    team2Condition = True
    if team1!= "all":
        team1Condition = ( (df.home == team1) | (df.away == team1))
    if team2!= "all":    
        team2Condition = ((df.home == team2) | (df.away == team2))
        
    return df[team1Condition & team2Condition & (abs(df.home_score.astype(int) - df.away_score.astype(int)) >= ecart)]
```

## Games with 6 Goals difference  between 2009 and 2019

```python
getScoreDifference(ecart=6)
```

##  PSG games with 5 Goals difference  between 2009 and 2019

```python
getScoreDifference(ecart=6, team1="Paris Saint-Germain")
```
